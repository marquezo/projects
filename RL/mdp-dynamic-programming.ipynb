## Markov Decision Processes and Dynamic Programming

### Comparing policy iteration, value iteration and modified policy iteration

Below I compare the performance of value iteration, policy iteration and modified policy iteration in terms of policy evaluation and policy improvement steps. Note that a policy improvement step is computationnally more expensive because a max takes place over all possible actions given a state. Depending on how many actions there are, this max operation can make a big difference. In all these cases, the discount factor is 0.95 and convergence means that the policy and value function are optimal to 0.001.

First, for the <b>simple 2-state MDP</b> (the initial policy is NOT the optimal policy in all these cases):

In this case, value iteration is the most wasteful because the optimal policy does not change (due to the very limited number of actions the agent can take). Thus, the max operation over actions is not needed but the iterations are still needed to compute the value of the policy to the desired precision. Modified policy iteration will find the optimal policy after 10 executions of policy evaluation (in above experiments, m=10 in all iterations) and thereafter the max operation over actions is useless but it performs fewer max operations than value iteration. In this case, policy iteration and modified policy iteration perform equally well because their number of executions of policy evaluation is comparable.

Now, for the <b>small Grid World</b> over 25 states. Here, the discount factor is 0.9 and convergence means that the policy and value function are optimal to 0.001.

Here, policy iteration and value iteration are both wasteful but policy iteration more so. In the case of policy iteration, multiple sweeps over the entire state set are not really required all the time, especially when the policy is not optimal. Value iteration wastes max operations over the action space. A quick inspection reveals that after 5 iterations, value iteration has already found an optimal policy whereas policy iteration finds an optimal policy after 2 policy improvements but after 148 policy evaluations. It is worth noting that the performance of policy iteration has the most variance: random initializations cause the number of iterations required for convergence to chance substantially.

Modified policy iteration finds an optimal policy after 4 iterations. It ends up doing only 11 max operations over the action space and only ends up doing 100 policy evaluations. Note that one could optimize the value of $m$ in this algorithm; in this case, I left it constant as 10.

As an aside, empirically one can verify that the one-array version of policy evaluation converges faster than the 2-array version, as mentioned in section 4.1 of the S&B book.

<h4>Changing the discount rate</h4>

To do this comparison, I only used experiments based on the Grid World problem. The graph at the bottom of this notebook shows that the convergence rate for the three algorithms is pretty similar for all values of discount factor except when this value is close to 1 (greater than 0.9). In this part of the graph, the time needed for value iteration to converge skyrockets. 

When the agent lives for a long time ($\gamma > 0.9$), that is, it lives for future rewards, the number of times that policy evaluation needs to take place to find a 0.001 optimal policy greatly increases. Modified policy iteration is most efficient because it evaluates the policy a lot but the max operation over the action space does not happen very often. On the other hand, value iteration will perform this max operation every time it evaluates the policy, which causes it to underperform as doing this max operation many times is not needed and is wasteful. Policy iteration performs too many policy evaluation operations but as it performs far less max operations than value iteration, it is worse than modified policy iteration but better than value iteration in general. 

However, the convergence rate of value iteration can be greatly improved by using only one-array in the solution (corresponding to the Gauss-Seidel implementation). This is also shown in the graph at the bottom of this notebook. In this case, Gauss-Seidel value iteration performs better than policy iteration but still worse than modified policy iteration.

In conclusion, from these experiments, it seems like modified policy iteration is the most efficient algorithm as it tries to reduce waste in computations, which seem inevitable in both value iteration and policy iteration.


```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
# Only two states and three actions available to the agent
class SimpleMDP:
    def __init__(self, numStates=2, discount_factor=0.95, theta=0.1):
        self.numStates = numStates
        self.theta = theta
        self.discount_factor = discount_factor
        self.V = np.random.rand(self.numStates)
        self.pi = np.zeros(2, dtype=np.int8)
        self.pi[0] = 1#np.random.randint(low=0, high=2)
        self.pi[1] = 2
        self.transition = np.zeros((3, 2, 2))
        #[action][from state][to state]
        self.transition[0][0][0] = 0.5
        # prob of going from state 0 to state 1
        self.transition[0][0][1] = 0.5
        self.transition[0][1][0] = 0
        self.transition[0][1][1] = 0 
        
        self.transition[1][0][0] = 0
        # prob of going from state 0 to state 1
        self.transition[1][0][1] = 1
        self.transition[1][1][0] = 0
        self.transition[1][1][1] = 0 
        
        self.transition[2][0][0] = 0
        # prob of going from state 0 to state 1
        self.transition[2][0][1] = 0
        self.transition[2][1][0] = 0
        self.transition[2][1][1] = 1 
        
    def reward(self, state, action):
        if state == 0:
            if action == 0:
                return 5
            elif action == 1:
                return 10
        elif state == 1:
            return -1
        
    #Return array of length states, where entry at i corresponds to the reward that pursuing actions[i]
    #would yield in state i
    def valid_actions(self, state):
        if state == 0:
            return np.array([0, 1])
        elif state == 1:
            # repeat the only valid action so that the array has the same size as the number of states
            return np.array([2, 2])          
        
    def evaluation(self):        
        delta = 100
        
        #Construct transition matrix for the current policy
        transition = np.zeros((2,2))                
        transition[0] = self.transition[self.pi[0]][0]
        transition[1] = self.transition[self.pi[1]][1]
  
        print 'Policy Iteration - Evaluation------------------------------------------'
        iter_idx = 0

        while delta >= self.theta:
            delta = 0
            curr_v = np.copy(self.V)
            rewards = np.array([self.reward(state=0, action=self.pi[0]), self.reward(1, self.pi[1])])
            self.V = rewards + self.discount_factor*curr_v.dot(transition.T)               
            delta = max(delta, np.abs(curr_v - self.V).max())
            iter_idx +=1
            print "V_{}: {}, V_{}: {}, delta: {}".format(iter_idx-1, curr_v, iter_idx, self.V, delta)
            
    def value_iteration(self):
        start_time = time.time()
        delta = 100
        transition = np.zeros((2,2))
        iter_idx = 0
        print "Running value iteration--------------------------------------------------"
        
        while delta >= self.theta:
            delta = 0
            curr_v = np.copy(self.V)
            
            for s in range(2):
                
                rewards = [ self.reward(s, a) for a in self.valid_actions(s) ]

                #first allowed action
                transition[0] = self.transition[self.valid_actions(s)[0]][s]
                #second allowed action
                transition[1] = self.transition[self.valid_actions(s)[1]][s]

                #Create array with V(s) for each allowed action; use it to pick the action with the larger V(s)
                values = rewards + self.discount_factor*curr_v.dot(transition.T)

                self.V[s] = values.max()            
            
            delta = max(delta, np.abs(curr_v - self.V).max())
            iter_idx +=1
            print "V_{}: {}, V_{}: {}, delta: {}".format(iter_idx-1, curr_v, iter_idx, self.V, delta)
                    
        #Come up with pi star
        for s in range(2):
            old_action = self.pi[s]
            
            rewards = [ self.reward(s, a) for a in self.valid_actions(s) ]
            
            #first allowed action
            transition[0] = self.transition[self.valid_actions(s)[0]][s]
            #second allowed action
            transition[1] = self.transition[self.valid_actions(s)[1]][s]
            
            #Create array with V(s) for each allowed action; use it to pick the action with the larger V(s)
            values = rewards + self.discount_factor*self.V.dot(transition.T)
            
            self.pi[s] = self.valid_actions(s)[np.argmax(values)]
                
        return time.time() - start_time
        
    def improvement(self):
        policy_stable = True
        transition = np.zeros((2,2))
        print "Policy Iteration - Improvement--------------------------------"
        
        for s in range(2):
            old_action = self.pi[s]
            
            rewards = [ self.reward(s, a) for a in self.valid_actions(s) ]
          
            #first allowed action
            transition[0] = self.transition[self.valid_actions(s)[0]][s]
            #second allowed action
            transition[1] = self.transition[self.valid_actions(s)[1]][s]
            
            #Create array with V(s) for each allowed action; use it to pick the action with the larger V(s)
            values = rewards + self.discount_factor*self.V.dot(transition.T)
            
            self.pi[s] = self.valid_actions(s)[np.argmax(values)]
            
            if old_action != self.pi[s]:
                policy_stable = False
                    
        return policy_stable
    
    def policyIteration(self):       
        start_time = time.time()
        policy_stable = False

        while policy_stable != True:    
            self.evaluation()
            policy_stable = self.improvement()

        return time.time() - start_time
    
    def modifiedPolicyIteration(self, num_evals=5):
        start_time = time.time()
        delta = 100
        transition = np.zeros((2,2)) 
        iter_idx = 0
        print "Running modified policy iteration--------------------------------------------------"        
        
        while True:
            curr_v = np.copy(self.V)
            
            #Step 2: Policy improvement and Step 3a
            for s in range(2):
                rewards = [ self.reward(s, a) for a in self.valid_actions(s) ]

                #first allowed action
                transition[0] = self.transition[self.valid_actions(s)[0]][s]
                #second allowed action
                transition[1] = self.transition[self.valid_actions(s)[1]][s]

                #Create array with V(s) for each allowed action; use it to pick the action with the larger V(s)
                values = rewards + self.discount_factor*curr_v.dot(transition.T)

                self.V[s] = values.max()
                self.pi[s] = self.valid_actions(s)[values.argmax()]                              
            
            delta = np.abs(curr_v - self.V).max()
            print self.pi
            
            if  delta < self.theta:
                return time.time() - start_time

            k = 0
            #Construct transition matrix for the current policy            
            transition[0] = self.transition[self.pi[0]][0]
            transition[1] = self.transition[self.pi[1]][1]
            rewards = np.array([self.reward(state=0, action=self.pi[0]), self.reward(1, self.pi[1])])

            #Do some policy evaluation
            for k in range(num_evals):                                
                self.V = rewards + self.discount_factor*self.V.dot(transition.T)
                
            iter_idx +=1
            print "V_{}: {}, V_{}: {}, delta: {}".format(iter_idx-1, curr_v, iter_idx, self.V, delta)
        
```

```python
policyIterationSimple = SimpleMDP(discount_factor=0.95, theta=0.001)
print "V(0): {}, pi(0): {}".format(policyIterationSimple.V, policyIterationSimple.pi)

time_taken = policyIterationSimple.policyIteration()

print "Policy iteration took {} seconds".format(time_taken)    
print "V*: {}, pi*: {}".format(policyIterationSimple.V, policyIterationSimple.pi)
```

### Grid World

```python
UP=0
DOWN=1
LEFT=2
RIGHT=3

class GridWorld:
    def __init__(self, discount_factor=0.95, theta=0.1, num_states=25):
        self.grid_transitions = np.loadtxt('grid_world_transitions.csv', dtype=np.int8).reshape(4,25,25)                
        self.discount_factor = discount_factor
        self.theta = theta
        self.num_states = num_states
        #init state-values randomly
        self.V = np.random.rand(self.num_states)
        #init agent's policy randomly
        self.pi = np.random.randint(low=0, high=4, size=self.num_states)    
    
    def reward(self, state, action):
        if state in (0,1,2,3,4) and action == UP:
            return -1
        elif state in (0,5,10,15,20) and action == LEFT:
            return -1
        elif state in (4,9,14,19,24) and action == RIGHT:
            return -1
        elif state in (20,21,22,23,24) and action == DOWN:
            return -1
        elif state == 1 and action == DOWN:
            return 10
        elif state == 3 and action == DOWN:
            return 5
        else:
            return 0
        
    def test_transitions(self):
        print np.nonzero(self.grid_transitions[:, 10])[1] == np.array([5 ,15, 10, 11])
        print np.nonzero(self.grid_transitions[:, 15])[1] == np.array([10, 20, 15, 16])
        print np.nonzero(self.grid_transitions[:, 1])[1] == np.array([1, 21, 0, 2])        
        
    def evaluation(self):        
        delta = 100
        
        #Construct transition matrix for the current policy
        transition = np.zeros((self.num_states, self.num_states), dtype=np.int8)
        rewards = np.zeros(self.num_states, dtype=np.int8)
        
        for state_idx in range(self.num_states):
            transition[state_idx] = self.grid_transitions[self.pi[state_idx]][state_idx]
            rewards[state_idx] = self.reward(state=state_idx, action=self.pi[state_idx])

        print 'Policy Iteration - Evaluation------------------------------------------'
        iter_idx = 0

        while delta >= self.theta:
            delta = 0
            curr_v = np.copy(self.V)            
            self.V = rewards + self.discount_factor*curr_v.dot(transition.T)   
            delta = max(delta, np.abs(curr_v - self.V).max())
            iter_idx +=1
            #print "V_{}: {}, V_{}: {}, delta: {}".format(iter_idx-1, curr_v, iter_idx, self.V, delta)
            print "iteration {}, delta: {}".format(iter_idx, delta)
            
    def improvement(self):
        policy_stable = True
        transition = np.zeros((self.num_states, self.num_states), dtype=np.int8)
        print "Policy Iteration - Improvement--------------------------------"
        
        for s in range(self.num_states):
            old_action = self.pi[s]
            
            rewards = [ self.reward(s, action) for action in range(0, 4) ]
            
            #Create array with V(s) for each allowed action; use it to pick the action with the larger V(s)
            values = rewards + self.discount_factor*self.V[np.nonzero(self.grid_transitions[:, s])[1]]
            
            self.pi[s] = values.argmax()
            
            if old_action != self.pi[s]:
                policy_stable = False
        
        print "pi: {}".format(self.pi)
        
        return policy_stable

    def policyIteration(self):       
        start_time = time.time()
        policy_stable = False

        while policy_stable != True:    
            self.evaluation()
            policy_stable = self.improvement()

        return time.time() - start_time
            
    def value_iteration(self, one_array=False):
        start_time = time.time()        
        delta = 100
        transition = np.zeros((self.num_states, self.num_states), dtype=np.int8)
        iter_idx = 0
        print "Running value iteration--------------------------------------------------"
        
        while delta >= self.theta:
            delta = 0
            curr_v = np.copy(self.V)
            pi_curr_iter = np.zeros(self.num_states, dtype=np.int8)
            
            for s in range(self.num_states):
                
                rewards = [ self.reward(s, action) for action in range(0, 4) ]                

                #Create array with V(s) for each allowed action; use it to pick the action with the larger V(s)
                #Note: If we change curr_v for self.V below, we get a huge improvement because we use
                #values as they are updated
                if one_array:
                    values = rewards + self.discount_factor*self.V[np.nonzero(self.grid_transitions[:, s])[1]]
                else:
                    values = rewards + self.discount_factor*curr_v[np.nonzero(self.grid_transitions[:, s])[1]]

                self.V[s] = values.max()  
                pi_curr_iter[s] = values.argmax()
            
            delta = max(delta, np.abs(curr_v - self.V).max())
            iter_idx +=1
            #print "V_{}: {}, V_{}: {}, delta: {}".format(iter_idx-1, curr_v, iter_idx, self.V, delta)
            print "iteration {}, delta: {}".format(iter_idx, delta)
            print "pi({}): {}".format(iter_idx, pi_curr_iter)
                    
        #Come up with pi star
        for s in range(self.num_states):
            old_action = self.pi[s]
            
            rewards = [ self.reward(s, action) for action in range(0, 4) ]                
            
            #Create array with V(s) for each allowed action; use it to pick the action with the larger V(s)
            values = rewards + self.discount_factor*self.V[np.nonzero(self.grid_transitions[:, s])[1]]
            
            self.pi[s] = values.argmax()
        
        return time.time() - start_time
            
    def modifiedPolicyIteration(self, num_evals=5):
        start_time = time.time()
        delta = 100
        transition = np.zeros((self.num_states, self.num_states), dtype=np.int8)    
        iter_idx = 0
        print "Running modified policy iteration--------------------------------------------------"        
        
        while True:
            curr_v = np.copy(self.V)
            
            #Policy improvement
            for s in range(self.num_states):
                rewards = [ self.reward(s, action) for action in range(0, 4) ]

                #Create array with V(s) for each allowed action; use it to pick the action with the larger V(s)
                #Note: If we change curr_v for self.V below, we get a huge improvement because we use
                #values as they are updated, but for comparison purposes, use curr_v
                values = rewards + self.discount_factor*curr_v[np.nonzero(self.grid_transitions[:, s])[1]]

                self.V[s] = values.max()
                self.pi[s] = values.argmax()
                
            delta = np.abs(curr_v - self.V).max() 
            
            if delta < self.theta:
                return time.time() - start_time

            k = 0
            
            #Construct transition matrix for the current policy
            transition = np.zeros((self.num_states, self.num_states), dtype=np.int8)
            rewards = np.zeros(self.num_states, dtype=np.int8)
        
            for state_idx in range(self.num_states):
                transition[state_idx] = self.grid_transitions[self.pi[state_idx]][state_idx]
                rewards[state_idx] = self.reward(state=state_idx, action=self.pi[state_idx])

            #Do some policy evaluation
            for k in range(num_evals):                                
                self.V = rewards + self.discount_factor*self.V.dot(transition.T)           
                
            iter_idx +=1            
            print "iteration {}, delta {}".format(iter_idx, delta)
            print "pi({}): {} ".format(iter_idx, self.pi)
                
```

### Exploring the effect of the discount factor on the convergence rate

```python
discount_factors = np.linspace(0.01, 0.99, 40)
times_pi = np.zeros(discount_factors.shape)
times_vi = np.zeros(discount_factors.shape)
times_vim = np.zeros(discount_factors.shape)
times_mpi = np.zeros(discount_factors.shape)

for idx, d_factor in enumerate(discount_factors):
    policyIterationGridWorld = GridWorld(discount_factor=d_factor, theta=0.001)
    times_pi[idx] = policyIterationGridWorld.policyIteration()
    
    valueIterationGridWorld = GridWorld(discount_factor=d_factor, theta=0.001)
    times_vi[idx] = valueIterationGridWorld.value_iteration()
    
    valueIterationGridWorld = GridWorld(discount_factor=d_factor, theta=0.001)
    times_vim[idx] = valueIterationGridWorld.value_iteration(one_array=True)
    
    mPolicyIterationGridWorld = GridWorld(discount_factor=d_factor, theta=0.001)
    times_mpi[idx] = mPolicyIterationGridWorld.modifiedPolicyIteration(num_evals=10)
    
```

```python
plt.figure(figsize=(20,6))
plt.plot(discount_factors, times_vi, label="Convergence for Value Iteration - two-array")
plt.plot(discount_factors, times_vim, label="Convergence for Value Iteration - one-array")
plt.plot(discount_factors, times_pi, label="Convergence for Policy Iteration")
plt.plot(discount_factors, times_mpi, label="Convergence for Modified Policy Iteration")

plt.xlabel('Discount factor')
plt.ylabel('Execution time (s)')
plt.legend()  
plt.show()
```
